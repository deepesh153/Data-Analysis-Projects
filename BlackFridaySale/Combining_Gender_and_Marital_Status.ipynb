```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('BlackFriday.csv')
# df.head()
```

```python
del df['Product_Category_2']
del df['Product_Category_3']
# # Deleting the columns
```

```python
l =[]
for i in range(len(df)):
    l.append((df['Gender'][i]+"_"+str(df['Marital_Status'][i])))
df['Marital_Gender'] = l    
```

```python
df.head()
```

```python
sns.countplot(x = df['Marital_Gender'])
```

### Making Age_Category

```python
l =[]
for i in range(len(df)):
    if(df['Age'][i]=='0-17'):
        l.append('Kids')
    elif(df['Age'][i]=='18-25' or df['Age'][i] == '26-35'):
        l.append('Adults')
    elif(df['Age'][i]=='36-45' or df['Age'][i] == '46-50'):
        l.append('MiddleAged')
    elif(df['Age'][i]=='51-55' or df['Age'][i] == '55+'):
        l.append('OldAged')
    else:
        break
df['Age_Category'] = l
```

```python
df.head()
```

```python
sns.countplot(x ='Age_Category',hue = 'Gender',data = df)
```
