```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

```python
df2 = df.select_dtypes(include =['number'])
df2.corr()
```

#### It's easier to find correlation using Heatmap

```python
sns.heatmap(df2.corr())
```

#### Negative Relations: MaxHR and HeartDisease , MaxHR and Age
#### Positive Relatiion: Oldpeak and HeartDisease(0.5)

### Using JointPlot

```python
sns.jointplot(x='HeartDisease',y ='Oldpeak',data = df,kind='hex')
```

#### No Relation

```python
sns.jointplot(x='Age',y ='MaxHR',data = df,kind='hex')
```

```python
sns.jointplot(x='Age',y ='MaxHR',data = df,kind='reg')
```

```python
sns.jointplot(x='Oldpeak',y ='MaxHR',data = df,kind='reg')#with regression line(flow of data)
```
