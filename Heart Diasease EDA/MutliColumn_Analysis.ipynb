```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

### Analysing Columns using Violin PLot

```python
sns.violinplot(x= df['HeartDisease'],y = df['Sex'])
```

#### The above graph shows that Males have more chances of HeartDisease while Females have very few.

```python
sns.violinplot(x= df['HeartDisease'],y = df['Age'])
```

#### Both of them are having eqaul but there is a slight difference:
#### Avg Age for being in the test is around 50 and Avg Age for being Positive is nearly 60.
#### Patients before 50 have less chance of being Positive.

```python
sns.violinplot(x= df['HeartDisease'],y = df['RestingBP'])
```

#### No Clear realtion

```python
sns.violinplot(x= df['HeartDisease'],y = df['Cholesterol']) # Here By Cholesterol Means Good Chlosterol.
```

#### Someone having Low Good Cholesterol is having chances of HeartDisease.

```python
sns.violinplot(x= df['HeartDisease'],y = df['ChestPainType'])
```

#### This shows that People with ATA and ASY Type of ChestPain have more Chances of Heart Disease.

```python
sns.violinplot(x= df['HeartDisease'],y = df['ExerciseAngina'])
```

#### People having ChestPain during Exercie have more chances of HeartDisease.

```python
sns.violinplot(x= df['HeartDisease'],y = df['FastingBS'])
```

#### People while Fast have less Chances of HeartDisease.

```python
sns.violinplot(x= df['HeartDisease'],y = df['ST_Slope'])
```

#### People with ST_Slope Up has less chances but with Flat and Down have high Chances.
