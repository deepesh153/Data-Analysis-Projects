```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('heart.csv')
df.head()
```

### Analysing Categorical Columns

### Pie Charts

```python
df.groupby('Sex').size().plot.pie(autopct = '%.1f' ,title = 'Patients by Sex')
```

#### 79% of the Patients are Male and 21% are Female.

```python
df.groupby('ChestPainType').size().plot.pie(autopct = '%.1f' ,title = 'Patients by ChestPainType')
```

#### More than Half(54%) of the Patients have ASY type of Chest Pain, then NAP(21%), then ATA(18%) , and atlast TA(5%).

```python
df.groupby('RestingECG').size().plot.pie(autopct = '%.1f' ,title = 'Patients by RestingECG')
```

#### Mostly Patients have Normal RestingECG and some of rest have LVH and ST.

```python
df.groupby('ST_Slope').size().plot.pie(autopct = '%.1f' ,title = 'Patients by ST_Slope')
```

#### Half of the Patients have Flat ST_Slope and 43% have Up and a few Patients have Down ST_Slope.

```python
df.groupby('HeartDisease').size().plot.pie(autopct = '%.1f' ,title = 'Patients by HeartDisease')
```

#### More than Half of the Patients have HeartDisease.
